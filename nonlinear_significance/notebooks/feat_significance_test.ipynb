```python
def linear_generative_model(n = 500,d = 10,sigma = 0.1,cor = 0.0):
    #n = 500
    #d = 10
    #sigma = 0.1
    Sigma = np.zeros((d, d))
    np.fill_diagonal(Sigma, 1)
    Sigma[0,1] = cor
    Sigma[1,0] = cor
    X = np.random.multivariate_normal(mean = np.zeros(d),cov = Sigma,size = n)
    #beta = np.random.randn(d)
    beta = np.zeros(d) 
    beta[0] = 1.0
    y = X @ beta + np.random.randn(n) * sigma
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = linear_generative_model(n = 1000,d = 10,sigma = 0.1,cor = 0.0)
```
