```python
def remove_elements(list1, list2):
    """
    Remove elements from list1 that are present in list2.
    
    Parameters:
    list1 (list): The original list.
    list2 (list): The list of elements to remove from list1.
    
    Returns:
    list: A new list with elements from list1, excluding those found in list2.
    """
    return [element for element in list1 if element not in list2]
```

```python
import matplotlib.pyplot as plt

# Define the number of ablation models and metrics
num_ablation_models = len(ablation_models[task])
num_metrics = len(metrics[task])

# Create a new figure with two columns for side-by-side plotting
fig, axs = plt.subplots(num_ablation_models, num_metrics * 2, figsize=(30, 20))

# Loop through the ablation models and metrics for correct prediction log loss
for i, a_model in enumerate(ablation_models[task]):
    for j, metric in enumerate(metrics[task]):
        results = {m: [] for m in methods_train_subset}
        
        # Calculate results for correct predictions
        for m in methods_train_subset:
            for k in range(num_features + 1):
                results[m].append(combined_df[combined_df['fi'] == m][a_model + f"_test_subset_correct_prediction_log_loss_after_ablation_{k}"].mean())
        
        ax = axs[i, j * 2]  # Adjust for side-by-side
        for m in methods_train_subset:
            color = color_map[m]
            linestyle = 'dashed' if m in ["TreeSHAP_RF", "Kernel_SHAP_RF_plus", "LIME_RF_plus", "Random"] else 'solid'
            ax.plot(range(num_features + 1), results[m], label=m, linestyle=linestyle, color=color)
        
        ax.set(xlabel='Number of features ablated', ylabel='Correct Prediction Metric',
               title=f'Ablation model = {a_model} (Correct)')
        if i == 0:
            ax.legend()

# Loop through the ablation models and metrics for incorrect prediction log loss
for i, a_model in enumerate(ablation_models[task]):
    for j, metric in enumerate(metrics[task]):
        results = {m: [] for m in methods_train_subset}
        
        # Calculate results for incorrect predictions
        for m in methods_train_subset:
            for k in range(num_features + 1):
                results[m].append(combined_df[combined_df['fi'] == m][a_model + f"_test_subset_incorrect_prediction_log_loss_after_ablation_{k}"].mean())
        
        ax = axs[i, j * 2 + 1]  # Adjust for side-by-side
        for m in methods_train_subset:
            color = color_map[m]
            linestyle = 'dashed' if m in ["TreeSHAP_RF", "Kernel_SHAP_RF_plus", "LIME_RF_plus", "Random"] else 'solid'
            ax.plot(range(num_features + 1), results[m], label=m, linestyle=linestyle, color=color)
        
        ax.set(xlabel='Number of features ablated', ylabel='Incorrect Prediction Metric',
               title=f'Ablation model = {a_model} (Incorrect)')
        if i == 0:
            ax.legend()

plt.tight_layout()
plt.savefig(f"./{task_name}_{task}_conditional_test_subset.png")
plt.show()
```
