```python
transform_ids = [361236, 361242, 361244, 361252, 361257, 361260, 361261, 361264, 361266, 361267, 361272, 361618, 361622]
```

```python
import numpy as np
```

```python
for id in transform_ids:
    task = openml.tasks.get_task(id, download_splits=False, download_data=False,
                                 download_features_meta_data=False,
                                 download_qualities=False)
    dataset = task.get_dataset()
    X, y, _, _ = dataset.get_data(target=dataset.default_target_attribute)
    print('Dataset ID:', id)
    print('Dataset Name:', dataset.name)
    print('Number of Features:', X.shape[1])
    print('Number of Instances:', X.shape[0])
    print('Histogram of Target Values:')
    plt.hist(y, bins=20)
    plt.show()
    # if min value of y is greater than 0, then log transform y
    if np.min(y) > 0:
        y = np.log(y)
    if np.min(y) <= 0:
        y = np.log(y - np.min(y) + 1)
    plt.hist(y, bins=20)
    plt.show()
    print('-----------------------------------')
```
